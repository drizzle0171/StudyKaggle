# 3.3 Feature Engineering Age

```python
df_train['Age_cat'] = 0
df_test['Age_cat'] = 0
```

```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
df_train.Age_cat = df_train.Age.apply(category_age)
df_test.Age_cat = df_test.Age.apply(category_age)
```

```python
df_train.drop(['Age'], axis = 1, inplace = True)
df_test.drop(['Age'], axis = 1, inplace = True)
```
